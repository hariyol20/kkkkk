# **Membaca Data**

Data yang digunakan adalah data tentang hubungan antara pendapatan dengan berbagai faktor diantaranya usia, jenis kelamin, pendidikan , jenis pekerjaan, dan lama bekerja.

```python
import pandas as pd

# Baca file CSV
data = pd.read_csv('Salary_Data.csv')
data["Salary"] = data["Salary"] / 10000
data.rename(columns={'Salary': 'Salary(million)'}, inplace=True)
data=data.head(500)

data

```

# **Grafik Line Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# draw lineplot
sns.lineplot(x="Years of Experience", y="Salary(million)", data=data)

# setting the title using Matplotlib
plt.title('Line plot Pendapatan dengan Pengalaman kerja')
plt.xlabel('Pengalaman Kerja(Tahun)')
plt.ylabel('Pendapatan(juta)')

plt.show()
```

Berdasarkan grafik line plot antara pengalaman kerja dengan pendapatan diatas informasi yang dapat di peroleh anatara lain:
1. Rata-rata pendapatan meningkat seiring dengan pengalaman kerja yang semakin lama. misalnya rata-rata pendapatan untuk pekerja dengan pengalaman kerja 5 tahun adalah sekitar 10 juta.
2. Tren pendapatan seiring bertambah dengan pengalaman kerja yang semakin meningkat. Hal ini bisa dilihat dari kurva yang menunjukan pendapatan semakin meningkat seiring dengan bertambahnya pengalaman kerja
3.Perbandingan pendapatan antara pekerja dengan tingkat pengalaman kerja yang berbeda. Misalnya, pendapatan pekerja dengan pengalaman kerja 10 tahun sekitar 2 kali lebih tinggi daripada pendapatan pekerja dengan pengalaman kerja 5 tahun.

# **Grafik Scatter Plot**

```python
from scipy.stats import pearsonr

# Menghapus baris yang mengandung nilai NaN
data_cleaned = data.dropna()

# Mengatur warna secara manual
color = {"Bachelor's": "blue", "Master's": 'yellow', 'PhD': 'red'}

# draw scatterplot
sns.scatterplot(x='Years of Experience', y='Salary(million)', data=data_cleaned, hue='Education Level', palette=color)
sns.regplot(x='Years of Experience', y='Salary(million)', data=data_cleaned, scatter=False)

# Hitung nilai korelasi Pearson
correlation_coefficient, p_value = pearsonr(data_cleaned['Years of Experience'], data_cleaned['Salary(million)'])

# Tampilkan nilai korelasi pada plot
correlation_text = f'Pearson Correlation: {correlation_coefficient:.2f}'
plt.text(0.5, 0.9, correlation_text, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10)

plt.title("Scatter plot Pengalaman lama bekerja dengan Pendapatan")
plt.ylabel("Pendapata(juta)")
plt.xlabel("Pengalaman bekerja(tahun)")
plt.show()
```

Berdasarkan grafik Scatter plot diatas, informasi yang dapat diperoleh adalah sebagai berikut:
1. Hubungan antara pengalaman kerja dan tingkat pendidikan dengan pendapatan  adalah linear positif. Hal ini berarti bahwa semakin lama pengalamanan kerja dan semakin tinggi tingkat pendidikan seseorang ,pendapatan yang akan diperoleh juga akan semakin tinggi.
2. Rata-rata pendapatan yang diperoleh dengan tingkat pendidikan sarjana dengan rentang pengalaman kerja antara 0-10 tahun berada pada kisaran 3 sampai 13 juta.
3. Pada tingkat pendidikan sarjana dan doktor terdapat seseorang yang memiliki pendapatan yang cukup tinggi diantara yang lainnya yaitu pada angka 25 juta dengan pengalaman kerja yang tentunya juga cukup lama yaitu 25 tahun.
4.Rata-rata pendapatan yang diperoleh oleh orang lulusan S2 lebih ditinggi daripada orang yang lulusan Sarjana dan lebih rendah dibandingkan dengan orang yang lulusan Doktor. Hal ini juga diikuti dengan pengalaman kerja yang mereka miliki.

# **Barplot**

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Mengatur gaya plot
sns.set_style("darkgrid")

# Membuat palet warna yang lebih menarik
custom_palette = sns.color_palette("colorblind",4)

# draw barplot dengan gaya dan palet warna baru
sns.barplot(x='Education Level', y='Salary(million)', data=data, hue='Gender', palette=custom_palette)

# Menambahkan nilai pada setiap batang
for p in plt.gca().patches:
    plt.annotate(format(p.get_height(), '.2f'),
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha = 'center', va = 'center',
                 xytext = (0, 10),
                 textcoords = 'offset points')

plt.title("Barplot Tingkat pendidikan dan Jenis kelamin dengan Pendapatan")
plt.ylabel("Pendapata(juta)")
plt.xlabel("Tingkat Pendidikan")

plt.show()
```

Berdasarkan grafik barplot diatas, informasi yang dapat diperoleh antara lain:
1. Rata-rata pendapatan yang diperoleh oleh laki-laki dan perempuan yang lulusan sarjana relatif sama yaitu laki-laki dengan pendapatan 9,15 juta dan perempuan dengan pendapatan 9,11 juta.
2. Pada lulusan S2(Master) terdapat perbedaan yang cukup signifikan antara pendapatan yang diperoleh oleh laki-laki dan perempuan yaitu pada angka 13,78 juta dan 11,60 juta.
3. Pada lulusan S3(Doktor), pendapatan yang diperoleh oleh laki-laki dan perempuan terddapat perbedaan yang tidak terlalu jauh yaitu pada angka 16,09 juta dan 15,87 juta.
4. Berdasarkan grafik tersebut dapat disimpulkan bahwa semakin tinggi pendidikan baik laki-laki maupun perempuan, pendapatan yang mereka peroleh juga semakin tinggi.
5. Berdasarkan grafik tersebut juga, pada setiap jenjang pendidikan,rata-rata laki-laki selalu mendominasi dengan pendapatan yang selalu lebih tinggi dibandingkan perempuan  

# **Histogram**

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#mengatur warna batang secara manual
color1= {"Bachelor's": "navy", "Master's": 'yellow', 'PhD': 'red'}

# draw barplot
sns.histplot(x='Salary(million)', data=data, kde = True, hue = 'Education Level',palette=color1)

plt.show()
```

Berdasarkan grafik tersebut, informasi yang bisa didapatkan adalah sebagai berikut:
1. Berdasarkan grafik tersebut data menjulur kekanan, yang menandakan bahwa semakin tinggi pendidikan seseorang, maka semakin tinggi pula pendapatan yang akan diperoleh
2.Rata-rata pendapatan untuk lulusan S1(Sarjana) berada pada rentang 2 sampai 10 juta
3. Rata-rata pendapatan lulusan S3(Doktor) lebih tinggi dibandikan dengan lulusan S2(Master) dan S1(Sarjana)

# **Diagram** **Gelembung**

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create PairGrid
plot = sns.PairGrid(data)
plot.map_upper(sns.scatterplot)  # Scatter plot di segitiga bagian atas
plot.map_lower(sns.scatterplot)  # Scatter plot di segitiga bagian bawah
plot.map_diag(plt.hist)  # Plot histogram di diagonal

plt.show()
```

2.Pada lulusan Sarjana dengan jenis kelamin laki-laki terdapat satu outlier/pencilan dengan pendapatan yang paling tinggi diantara yang lainnya yaitu pada angka 25 juta.
3. Pada lulusan Master(S2) data pada jenis kelamin laki-laki dan perempuan sama-sama menjulur kekanan dengan mediannya berada pada 13 juta dan 10 juta
